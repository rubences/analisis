Paso 1: Definir un Objetivo para el Estudio
Objetivo del Estudio:
Supongamos que elegimos una base de datos relacionada con el rendimiento académico de estudiantes en una universidad. El objetivo podría ser "Identificar los factores que influyen significativamente en el rendimiento académico de los estudiantes".

Variables y su Significado:

GPA (Variable dependiente): El promedio de calificaciones, que indica el rendimiento académico.
Horas de Estudio (Variable independiente): Cantidad de horas que un estudiante dedica al estudio por semana.
Participación en Actividades Extraescolares (Variable independiente): Si el estudiante participa (1) o no (0) en actividades extraescolares.
Asistencia a Clases (Variable independiente): Porcentaje de asistencia a clases.
Objeto de Estudio:
Analizar cómo las horas de estudio, la participación en actividades extraescolares y la asistencia a clases influyen en el GPA de los estudiantes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats

# Generar datos de muestra
np.random.seed(0)
data = {
    "GPA": np.random.normal(3.0, 0.5, 100),
    "Study_Hours": np.random.normal(15, 5, 100),
    "Extra_Curricular": np.random.choice([0, 1], 100),
    "Class_Attendance": np.random.uniform(70, 100, 100)
}

df = pd.DataFrame(data)

# Paso 2: Exploración de Datos
# Histograma de GPA
plt.figure(figsize=(8, 5))
sns.histplot(df['GPA'], kde=True)
plt.title('Distribución de GPA')
plt.xlabel('GPA')
plt.ylabel('Frecuencia')
plt.show()

# Gráfico de dispersión entre GPA y Horas de Estudio
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Study_Hours', y='GPA', data=df)
plt.title('GPA vs Horas de Estudio')
plt.xlabel('Horas de Estudio')
plt.ylabel('GPA')
plt.show()

# Diagrama de cajas por Participación en Actividades Extraescolares
plt.figure(figsize=(8, 5))
sns.boxplot(x='Extra_Curricular', y='GPA', data=df)
plt.title('GPA por Participación en Actividades Extraescolares')
plt.xlabel('Participación en Actividades Extraescolares (0 = No, 1 = Sí)')
plt.ylabel('GPA')
plt.show()

# Gráfico de dispersión entre Asistencia a Clases y GPA
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Class_Attendance', y='GPA', data=df)
plt.title('GPA vs Asistencia a Clases')
plt.xlabel('Asistencia a Clases (%)')
plt.ylabel('GPA')
plt.show()

# Paso 3: Análisis Estadístico
# Correlación y Regresión
X = df[['Study_Hours', 'Extra_Curricular', 'Class_Attendance']]
X = sm.add_constant(X)  # Añadir constante
y = df['GPA']
model = sm.OLS(y, X).fit()
print(model.summary())

# Prueba T de Estudiantes para Actividades Extraescolares
ttest_result = stats.ttest_ind(
    df[df['Extra_Curricular'] == 1]['GPA'],
    df[df['Extra_Curricular'] == 0]['GPA']
)
print("Prueba T de Estudiantes para Actividades Extraescolares:\n", ttest_result)

# Resumen de resultados
# Agregar aquí las interpretaciones y conclusiones de los análisis realizados
```
